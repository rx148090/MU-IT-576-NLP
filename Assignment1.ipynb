```python
#nltk.download('inaugural')
#nltk.download('stopwords')
from sklearn.linear_model import LinearRegression
import string
from nltk.corpus import inaugural, stopwords
stop_words = set(stopwords.words('english'))
```

```python
# create the vocabulary for the corpus
vocab = []
for fileid in inaugural.fileids():
    lst = inaugural.words(fileids=[fileid])
    # remove stopwords and puctuations
    words = [word.lower() for word in lst if word.lower() not in stop_words and word not in string.punctuation]
    vocab.extend(word for word in words if word not in vocab)
vocab[:10]
```

```python
# create the conditional frequency dictionaries
cfd = nltk.ConditionalFreqDist((target, fileid[:4])for fileid in inaugural.fileids()for w in inaugural.words(fileid)for target in vocab if w.lower().startswith(target))
```

```python
# split at year 2009
Xs = [fileid[:4] for fileid in inaugural.fileids()]
bf_2009 = Xs[:Xs.index('2009')]
af_2009 = Xs[Xs.index('2009'):]
```
